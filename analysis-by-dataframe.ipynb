```python
import re

def reading_source(path):
    list = []
    listv = {}
    sourcecode = open(path,"r+", encoding='utf-8')
    data=sourcecode.readlines()
    condition = 0
    for idata in data :

        if idata[0] =="[" :
            name = idata.split(']')
            name = name[0][1:]
            print(name)
            if name not in list:
                print(name + " 추가되었습니다.")
                list.append(name)
                listv[name]= 1
            else :
                listv[name] = listv[name] +1
    for i in listv :
        print(i +"\t" + str(listv[i]))


    print(listv)

if __name__ == '__main__':
    mylist = r"/Users/choihyeoktae/Desktop/chat-sample.csv"

    reading_source(mylist)
```
